```python
import pandas as pd
from cfa import cfa_single_layer, performance_plot, RSC_graph, compute_cd_ds, compute_performance
```

```python
# load df data
df = pd.read_csv('test_df.csv')
df.head(3)
```

```python
# load target vector
y_true = pd.read_csv('test_ytrue.csv')
y_true.head(3)
```

```python
# draw rank-score function
RSC_graph(df)
```

```python
# run cfa, with performance metric = accuracy or auroc
fusion_df = cfa_single_layer(df, y_true, perf_metric='accuracy')
fusion_df
```

```python
# performance plot and return sorted fusion df, so you can draw it based on different plot settings
fusion_df_sorted = performance_plot(fusion_df, sort_col='asc')
fusion_df_sorted
```
